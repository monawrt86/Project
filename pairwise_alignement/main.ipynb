```python
import os
from Bio import SeqIO
from Bio.Seq import Seq
from Bio.SeqRecord import SeqRecord
from transformers import T5Tokenizer, T5EncoderModel
from utils import prottrans_T5_features
from funcs import local_pairwise_alignment

# Affiche la version de PyTorch pour le débogage
print(torch.__version__)

# Définit l'appareil de calcul sur le CPU
device = 'cpu'

# Nom du modèle T5 pré-entraîné
prot_model_name = 'Rostlab/prot_t5_xl_uniref50'

# Chargement du tokenizer et du modèle
tokenizer = T5Tokenizer.from_pretrained(prot_model_name, legacy=True)
model = T5EncoderModel.from_pretrained(prot_model_name).to(device)

# Définit la pénalité pour les gaps
gap_penalty = 1

# Définit le répertoire de sortie
output_dir = 'C:\\Users\\wrtmo\\Documents\\Internship\\pairwise_alignment\\data\\homstrad_data\\'
if not os.path.exists(output_dir):
    os.makedirs(output_dir)

# Charger les séquences dans un dictionnaire
sequences_dict = {record.id: str(record.seq) for record in SeqIO.parse(seqfa, "fasta")}  # Assurez-vous que le chemin "seqfa" est correct

# Traite chaque paire du fichier pairs.dat
with open(pairs, "r") as file:
    for line in file:
        ids = line.strip().split()
        if len(ids) == 2:
            id1, id2 = ids
            seq1 = sequences_dict.get(id1, "Séquence non trouvée")
            seq2 = sequences_dict.get(id2, "Séquence non trouvée")

            # Extraction des caractéristiques
            features = prottrans_T5_features([seq1, seq2], tokenizer, model, device)

            # Sélectionne les caractéristiques pour chaque séquence
            f1 = features[0, :len(seq1)]
            f2 = features[1, :len(seq2)]

            # Effectue l'alignement par paires
            aligned_seq1, aligned_seq2 = local_pairwise_alignment(seq1, seq2, f1, f2, gap_penalty)

            # Crée les objets SeqRecord
            r1 = SeqRecord(Seq(aligned_seq1.replace('\n', '').replace('\r', '')), id=id1, description='')
            r2 = SeqRecord(Seq(aligned_seq2.replace('\n', '').replace('\r', '')), id=id2, description='')

            # Chemin de sortie pour le fichier FASTA spécifique à la paire
            output_path = os.path.join(output_dir, f"{id1}_{id2}.fasta")

            # Écriture des séquences alignées dans le fichier FASTA
            SeqIO.write([r1, r2], output_path, 'fasta')

            # Confirmation de la fin de traitement pour la paire
            print(f"Alignement sauvegardé pour {id1} et {id2} dans {output_path}")
```

```python
# Importation des modules nécessaires
import os  # Module pour les opérations sur le système de fichiers
from Bio import SeqIO  # Module de Biopython pour lire les fichiers FASTA
from transformers import T5Tokenizer, T5EncoderModel  # Modules de Transformers pour le modèle T5
import torch  # Module pour les opérations sur les tenseurs

# Fonction pour extraire les paires d'un alignement
def pairs_from_aln(start1, aln1, start2, aln2):
    i1, i2 = start1, start2  # Initialisation des positions de départ
    pairs = []  # Liste pour stocker les paires d'indices
    for letter1, letter2 in zip(aln1, aln2):  # Parcours simultané des deux alignements
        if letter1 not in ('-', '/') and letter2 not in ('-', '/'):  # Vérifie que les lettres ne sont pas des gaps
            pairs.append((i1, i2))  # Ajoute la paire (i1, i2) à la liste
        if letter1 not in ('-', '/'):  # Si letter1 n'est pas un gap
            i1 += 1  # Incrémente l'indice i1
        if letter2 not in ('-', '/'):  # Si letter2 n'est pas un gap
            i2 += 1  # Incrémente l'indice i2
    return pairs  # Retourne la liste des paires

# Fonction pour lire les alignements d'un fichier FASTA
def read_alignment(filepath):
    records = list(SeqIO.parse(filepath, "fasta"))  # Lecture du fichier FASTA et stockage des séquences
    return records[0].id, str(records[0].seq), records[1].id, str(records[1].seq)  # Retourne les IDs et séquences

# Fonction pour extraire les caractéristiques de ProtT5
def prottrans_T5_features(sequences, tokenizer, model, device):
    inputs = tokenizer(sequences, return_tensors='pt', padding=True, truncation=True).to(device)  # Tokenisation des séquences
    with torch.no_grad():  # Désactivation de la rétropropagation pour économiser de la mémoire
        outputs = model(**inputs)  # Exécution du modèle pour obtenir les caractéristiques
    features = outputs.last_hidden_state.cpu().numpy()  # Conversion des caractéristiques en numpy array
    return features  # Retourne les caractéristiques

# Initialisation du modèle T5
device = 'cpu'  # Utilisation du CPU (peut être changé en 'cuda' pour GPU)
prot_model_name = 'Rostlab/prot_t5_xl_uniref50'  # Nom du modèle pré-entraîné
tokenizer = T5Tokenizer.from_pretrained(prot_model_name, legacy=True)  # Chargement du tokenizer pré-entraîné
model = T5EncoderModel.from_pretrained(prot_model_name).to(device)  # Chargement du modèle pré-entraîné et déplacement sur le device

# Chemin correct vers le fichier de référence
reference_file = 'C:\\Users\\wrtmo\\Documents\\Internship\\pairwise_alignement\\output_refalign.fasta'

# Lire les alignements de référence
reference_alignments = {}  # Dictionnaire pour stocker les alignements de référence
with open(reference_file, 'r') as file:  # Ouverture du fichier en lecture
    lines = file.readlines()  # Lecture de toutes les lignes du fichier
    i = 0  # Initialisation de l'index
    while i < len(lines):  # Boucle sur les lignes du fichier
        header = lines[i].strip()  # Lecture de l'en-tête
        if header.startswith('>'):  # Vérifie si la ligne est un en-tête de séquence
            parts = header.split()  # Split de l'en-tête en parties
            if len(parts) >= 2:  # Vérifie qu'il y a au moins deux parties
                name1 = parts[0][1:]  # Récupère le premier nom sans le '>'
                name2 = parts[1]  # Récupère le deuxième nom
                seq1 = lines[i+1].strip()  # Lecture de la première séquence
                seq2 = lines[i+2].strip()  # Lecture de la deuxième séquence
                reference_alignments[(name1, name2)] = (seq1, seq2)  # Stockage des alignements dans le dictionnaire
                i += 2  # Incrémente l'index pour sauter à la prochaine séquence
        i += 1  # Incrémente l'index pour la prochaine ligne
print('oki')  # Indication que le traitement est terminé

# Créer les dossiers pour chaque caractéristique
output_base_dir = "scores"  # Répertoire de base pour les scores
os.makedirs(output_base_dir, exist_ok=True)  # Création du répertoire s'il n'existe pas déjà
for i in range(1, 1025):  # Boucle de 1 à 1024 inclus
    os.makedirs(os.path.join(output_base_dir, f"score_feature{i}"), exist_ok=True)  # Création de sous-répertoires pour chaque caractéristique
```

```python
# Importation des modules nécessaires
import os  # Module pour les opérations sur le système de fichiers
from Bio import SeqIO  # Module de Biopython pour lire les fichiers FASTA
from Bio.Seq import Seq  # Module pour les séquences biologiques
from Bio.SeqRecord import SeqRecord  # Module pour les enregistrements de séquences
from transformers import T5Tokenizer, T5EncoderModel  # Modules de Transformers pour le modèle T5
import torch  # Module pour les opérations sur les tenseurs
from utils import prottrans_T5_features  # Import de la fonction d'extraction de caractéristiques (à vérifier)
from funcs import local_pairwise_alignment  # Import de la fonction d'alignement par paires locales (à vérifier)

# Lire les alignements des fichiers du dossier "result"
result_dir = 'C:\\Users\\wrtmo\\Documents\\Internship\\pairwise_alignement\\result'  # Chemin du répertoire contenant les résultats
result_files = [f for f in os.listdir(result_dir) if os.path.isfile(os.path.join(result_dir, f))]  # Liste des fichiers dans le répertoire

# Affichage du nombre de fichiers trouvés
print(f"Fichiers trouvés dans le dossier 'result': {len(result_files)}")

# Boucle sur chaque fichier de résultats
for result_file in result_files:
    filepath = os.path.join(result_dir, result_file)  # Chemin complet du fichier
    name1, aln1, name2, aln2 = read_alignment(filepath)  # Lecture des alignements à partir du fichier
    
    # Affichage des informations sur le fichier et les séquences
    print(f"Traitement du fichier: {result_file}, Séquences: {name1}, {name2}")
    
    ref_aln = reference_alignments.get((name1, name2))  # Récupération de l'alignement de référence
    if ref_aln:  # Si l'alignement de référence existe
        start1, start2 = 0, 0  # Initialisation des positions de départ
        ref_pairs = pairs_from_aln(start1, ref_aln[0], start2, ref_aln[1])  # Extraction des paires de référence
        pairs = pairs_from_aln(start1, aln1, start2, aln2)  # Extraction des paires de l'alignement courant

        if ref_pairs:  # Calcul de la sensibilité si des paires de référence existent
            sensitivity = len([p for p in ref_pairs if p in pairs]) / len(ref_pairs)
        else:
            sensitivity = 0

        if pairs:  # Calcul de la précision si des paires courantes existent
            precision = len([p for p in pairs if p in ref_pairs]) / len(pairs)
        else:
            precision = 0

        # Affichage de la sensibilité et de la précision
        print(f"Sensitivity: {sensitivity:.4f}, Precision: {precision:.4f}")

        # Calcul des caractéristiques des séquences
        seq1, seq2 = aln1.replace('-', ''), aln2.replace('-', '')  # Suppression des gaps dans les séquences
        features = prottrans_T5_features([seq1, seq2], tokenizer, model, device)  # Extraction des caractéristiques
        print(f"Features extraites pour les séquences {name1} et {name2}")

        # Boucle sur chaque caractéristique
        for feature_index in range(1, 1025):
            f1 = features[0, :len(seq1), feature_index-1]  # Extraction de la caractéristique pour la première séquence
            f2 = features[1, :len(seq2), feature_index-1]  # Extraction de la caractéristique pour la deuxième séquence
            
            # Effectuer l'alignement par paires pour la caractéristique spécifique
            aligned_seq1, aligned_seq2 = local_pairwise_alignment(seq1, seq2, f1, f2, gap_penalty=1)
            if aligned_seq1 and aligned_seq2:  # Si l'alignement est réussi
                specific_pairs = pairs_from_aln(start1, aligned_seq1, start2, aligned_seq2)  # Extraction des paires spécifiques
                
                if ref_pairs:  # Calcul de la sensibilité spécifique si des paires de référence existent
                    specific_sensitivity = len([p for p in ref_pairs if p in specific_pairs]) / len(ref_pairs)
                else:
                    specific_sensitivity = 0

                if specific_pairs:  # Calcul de la précision spécifique si des paires spécifiques existent
                    specific_precision = len([p for p in specific_pairs if p in ref_pairs]) / len(specific_pairs)
                else:
                    specific_precision = 0
                
                # Création du chemin de sortie pour les résultats de la caractéristique
                feature_dir = os.path.join(output_base_dir, f"score_feature{feature_index}")
                output_filepath = os.path.join(feature_dir, f"{name1}_{name2}.txt")
                
                # Écriture des résultats dans un fichier
                with open(output_filepath, 'w') as output_file:
                    output_file.write(f"Sensitivity: {specific_sensitivity:.4f}\n")
                    output_file.write(f"Precision: {specific_precision:.4f}\n")
                    
                # Affichage de l'écriture du fichier
                print(f"Écrit dans le fichier: {output_filepath}")
    
    else:  # Si l'alignement de référence n'est pas trouvé
        print(f"Alignement de référence non trouvé pour: {name1}, {name2}")

    # Indication que le traitement du fichier est terminé
    print(f"Terminé: {result_file}")

# Indication que le traitement complet est terminé
print("Traitement complet.")
```
